```python

MD_x = data.iloc[:, 0:11]

# Converting 'Yes' to 1 and others to 0
MD_x = (MD_x == 'Yes').astype(int)

# Calculating column means and rounding them to 2 decimal places
col_means = MD_x.mean().round(2)

# Display the column means
print(col_means)
```

```python
MD_x
```

```python
from sklearn.decomposition import PCA

# Assuming MD_x DataFrame is already prepared (binary matrix)
# Step 1: Perform PCA
pca = PCA(n_components=11)
MD_pca = pca.fit(MD_x)

# Step 2: Summary of PCA results

# Standard deviation (square root of eigenvalues)
std_devs = np.sqrt(pca.explained_variance_)

# Proportion of variance explained by each component
prop_variance = pca.explained_variance_ratio_

# Cumulative proportion of variance
cum_variance = np.cumsum(prop_variance)

# Display PCA results
print("Standard deviation:", np.round(std_devs, 4))
print("Proportion of Variance:", np.round(prop_variance, 4))
print("Cumulative Proportion:", np.round(cum_variance, 4))

# Step 3: Factor loadings (eigenvectors or rotation matrix)
# This is equivalent to R's rotation matrix
loadings = pca.components_.T
loadings_df = pd.DataFrame(loadings, columns=[f'PC{i+1}' for i in range(11)], index=MD_x.columns)

print("\nFactor Loadings (Rotation Matrix):")
print(loadings_df.round(2))

# Step 4: Project data onto the principal component space
MD_pca_projection = pca.transform(MD_x)

# Plotting the projection in the first two principal components (PC1 vs PC2)
plt.figure(figsize=(8, 6))
plt.scatter(MD_pca_projection[:, 0], MD_pca_projection[:, 1], color='grey', alpha=0.5)
plt.title("PCA: Projection onto the First Two Principal Components")
plt.xlabel('PC1')
plt.ylabel('PC2')

# Add arrows representing the original variables in the new space
for i, var in enumerate(MD_x.columns):
    plt.arrow(0, 0, loadings[i, 0]*5, loadings[i, 1]*5, color='r', alpha=0.8)
    plt.text(loadings[i, 0]*5.2, loadings[i, 1]*5.2, var, color='r')

plt.grid()
plt.show()
```

```python
MD_x['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Female' else 0)
```

```python
# Define a mapping for the 'VisitFrequency' column
visit_freq_mapping = {
    'Every three months': 1/90,
    'Once a week': 1/7,
    'Once a month': 1/30,
    'Once a year': 1/365,
    'More than once a week': 2/7,  # Assume twice a week
    'Never': 1/3650  # Assume very infrequent, like once in ten years
}

# Apply the mapping to the 'VisitFrequency' column
MD_x['VisitFrequency'] = data['VisitFrequency'].map(visit_freq_mapping)


```

```python
def convert_likes(like_value):
    if 'I love it!' in like_value:
        return 5  
    elif 'I hate it!' in like_value:
        return -5 
    else:
        return int(like_value)

# Apply the conversion function to the 'likes' column
MD_x['Like'] = data['Like'].apply(convert_likes)

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# Assuming MD_x is your DataFrame
# Selecting only the relevant features for clustering
features = MD_x[['yummy', 'convenient', 'spicy', 'fattening', 
                 'greasy', 'fast', 'cheap', 'tasty', 
                 'expensive', 'healthy', 'disgusting','Like']]

# Normalize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Step 1: Elbow Method to find the best k
inertia = []
silhouette_scores = []

# Testing different values of k
k_values = range(2, 11)  # You can adjust the range if needed

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    
    # Calculate inertia
    inertia.append(kmeans.inertia_)
    
    # Calculate silhouette score
    if k > 1:  # Silhouette score is defined for k=2 and above
        score = silhouette_score(features_scaled, kmeans.labels_)
        silhouette_scores.append(score)

# Print the lengths of inertia and silhouette_scores for debugging
print("Length of inertia:", len(inertia))
print("Length of silhouette_scores:", len(silhouette_scores))

# Plotting Elbow Method
plt.figure(figsize=(12, 5))

# Inertia plot
plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')

# Silhouette Score plot
plt.subplot(1, 2, 2)
# Adjusted line to ensure correct lengths
plt.plot(range(2, len(silhouette_scores) + 2), silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Score')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')

plt.tight_layout()
plt.show()

# Step 2: Choose the best k based on the plots
best_k = 4  # Replace this with the optimal k based on the plots

# Step 3: Apply KMeans clustering with the best k
kmeans_final = KMeans(n_clusters=best_k, random_state=42)
MD_x['kmeans_labels'] = kmeans_final.fit_predict(features_scaled)

# Display the first few rows with cluster labels
print(MD_x.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Assuming MD_x is your DataFrame and you have selected relevant features
features = MD_x[['yummy', 'convenient', 'spicy', 'fattening', 
                 'greasy', 'fast', 'cheap', 'tasty', 
                 'expensive', 'healthy', 'disgusting','Like']]

# Range of k values to test
k_values = range(1, 11)  # You can adjust the range if needed
bics = []

# Fit GMM for each k and calculate BIC
for k in k_values:
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm.fit(features)
    bics.append(gmm.bic(features))

# Plot BIC values against k values
plt.figure(figsize=(8, 4))
plt.plot(k_values, bics, marker='o')
plt.title('BIC vs Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('BIC')
plt.xticks(k_values)
plt.grid()
plt.show()

# Determine the best k (the one with the lowest BIC)
best_k = k_values[np.argmin(bics)]
print(f'The best number of clusters (k) is: {best_k}')
```

```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Define the independent variables (leaving out irrelevant columns)
independent_vars = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 
                    'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting']

# Define dependent variable
dependent_var = 'Like'

# Step 1: Fit Gaussian Mixture Model (GMM) for latent class clustering
gmm = GaussianMixture(n_components=4, random_state=1234)
MD_x['gmm_labels'] = gmm.fit_predict(MD_x[independent_vars])

# Step 2: Fit Linear Regression for Each Cluster (latent class regression equivalent)
def fit_latent_class_regression(data, independent_vars, dependent_var, n_components=4):
    # Perform Gaussian Mixture Modeling (clustering)
    gmm = GaussianMixture(n_components=n_components, random_state=1234)
    
    # Fit GMM on the independent variables
    gmm.fit(data[independent_vars])
    
    # Assign clusters to each row
    data['cluster'] = gmm.predict(data[independent_vars])
    
    # Fit linear regression models for each cluster and print coefficients
    for cluster in range(n_components):
        cluster_data = data[data['cluster'] == cluster]
        X = cluster_data[independent_vars]  # Select the independent variables
        y = cluster_data[dependent_var]  # Dependent variable (Like)
        
        reg_model = LinearRegression()
        reg_model.fit(X, y)
        
        print(f"Regression coefficients for cluster {cluster + 1}:")
        print(reg_model.coef_)
        print("Intercept:", reg_model.intercept_)
        
        # Plot the regression coefficients for the current cluster
        plot_regression_coefficients(reg_model, independent_vars, cluster)

# Step 3: Helper function to plot coefficients
def plot_regression_coefficients(model, feature_names, cluster_num):
    coefficients = model.coef_
    plt.barh(feature_names, coefficients)
    plt.title(f"Regression Coefficients for Cluster {cluster_num + 1}")
    plt.xlabel("Coefficient Value")
    plt.ylabel("Features")
    plt.show()

# Step 4: Fit latent class regression and plot for 4 clusters
fit_latent_class_regression(MD_x, independent_vars, dependent_var, n_components=4)
```
